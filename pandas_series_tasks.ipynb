###`Q-2:` Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

````python
# code here
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 9])

addition  =  series1 + series2
subtract  =  series1 - series2
multiple  =  series1 * series2
divide  =  series1 / series2

print(addition)
print(subtract)
print(multiple)
print(divide)
````

###`Q-3` Write a Pandas program to compare the elements of the two Pandas Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 10]



````python
# code here
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 10])

equal = series1 == series2
greater_than = series1 > series2
less_than = series1 < series2

print(equal)
print(greater_than)
print(less_than)
````

###`Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
# code here
def change_dtype(series,dtype):
  return pd.to_numeric(series,errors = "coerce", downcast=dtype)

series = pd.Series([1,2, "Python",2.0, True, 100])
print(change_dtype(series,"float"))
````

###`Q-9`
    i. Read `items.csv` making `item_name` as index.
    ii. Show no of nan values
    ii. Item price is given in $, so convert it to rupees without currency symbol.
    iii. Make data type of newly made series as float.
    iv. Fill nan with mean of the series


How csv file looks

```
item_name	item_price
Chips and Fresh Tomato Salsa	$2.39
Izze	$3.39
Nantucket Nectar	$3.39
Chips and Tomatillo-Green Chili Salsa	$2.39
Chicken Bowl	$16.98

```

````python
# code here
items = pd.read_csv("/content/items.csv",index_col="item_name").iloc[:,0]
print("no of nan values = ",items.isna().sum())
price_rupees = items.apply(lambda x : float(x.replace("$",""))*280 if pd.notna(x) else x)
price_rupees.astype(float)
price_rupees = price_rupees.fillna(price_rupees.mean())
````

````python
price_rupees
````

###`Q-10`:
    i. Find mean price
    ii. Find 30th and 6th percentile value
    iii. Plot Histogram on price with bin size 50
    iv. No of items price lies between [1000 to 2000]



````python
# code here
price_mean = items.apply(lambda x : float(x.replace("$","")) if pd.notna(x) else x).mean()
print(price_mean)
percentile = items.apply(lambda x : float(x.replace("$","")) if pd.notna(x) else x).quantile([0.30,0.06])
print(percentile)
print(items.hist(bins=50))
print(price_rupees[price_rupees.between(1000,2000)].size)
````
